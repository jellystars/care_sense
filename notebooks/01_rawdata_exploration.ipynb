# data exploration
Initial exploration of the CMS Hospital Readmissions Reduction Program dataset.

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# build path to notebook so it can work on any machine
RAW_DATA = Path("../data/raw/readmissions.csv")

# make sure the file exists before trying to load it
if not RAW_DATA.exists():
    raise FileNotFoundError(
        f"Expected file not found: {RAW_DATA}\n"
        "Download from https://data.cms.gov/provider-data/dataset/9n3s-kdb3\n"
        "Rename to readmissions.csv and place in data/raw/"
    )

print(f"File found: {RAW_DATA}")
```

```python
# load the data
df = pd.read_csv(RAW_DATA, dtype=str) # will clean types later

print("Dataset succesfully loaded")
```

```python
rows, cols = df.shape
print(f"Rows: {rows:,}")
print(f"Columns: {cols}")
```

```python
print("Data types:\n")
print(df.dtypes.to_string())
```

```python
df.head()
```

```python
# check for missing values
missing = (
    df.isnull()
    .sum()
    .reset_index()
    .rename(columns={"index": "column", 0: "missing_count"})
)

# find the percentage of missing values for each row
missing["missing_percentage"] = (missing["missing_count"] / len(df) * 100).round(2)

# focus on just the missing values
missing = missing[missing["missing_count"] > 0].sort_values("missing_count", ascending=False)

if missing.empty:
    print("No missing values found.")
else:
    print(f"Columns with missing values: {len(missing)}\n")
    print(missing.to_string(index=False))
```
